```python
# dividing the data into attributes and label which is x , y  which are also MaxTemp, MinTemp respectively
```

```python
x = dataset['MinTemp'].values.reshape(-1, 1)
y = dataset['MaxTemp'].values.reshape(-1, 1)
```

```python
# We slit 80% of the  data into training set
# 20% into test using scikit-learn
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.2, random_state=0)
```

```python
# after splitting of data 
# WE Train our Algorithm
```

```python
regressor = LinearRegression()
regressor.fit(x_train, y_train) # this is the training we're talking about
```

```python
# to get the intercept:
print(regressor.intercept_[0])

# to get the slope:
print(regressor.coef_[0][0])
```
